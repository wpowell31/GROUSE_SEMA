```python
import pandas as pd
import numpy as np

df_weight = pd.read_csv('df_weight')
df_weight.head()
```

```python
import statsmodels.api as sm
import numpy as np
import warnings
#warnings.filterwarnings("ignore")


#Clean the data to include only the SHAP features
shap_features = ['SEX', 'SEMA_WEEKS', 'dulaglutide', 'atorvastatin', 'AGE', 'metformin', 'potassium chloride',
                 'formoterol', 'lisdexamfetamine', 'INIT_WT', 'insulin aspart, human', 'triamcinolone',
                 'bupropion', 'sugammadex', 'pantoprazole', 'fenofibrate', 'ENDO', 'insulin lispro',
                 'methylprednisolone', 'lorazepam', 'Label']

df_weight = df_weight[shap_features]

#Splitting into predictor and label categories
X = df_weight.iloc[:,:-1]
y = [1 if a<-.1 else 0 for a in df_weight.iloc[:,-1]]

model = sm.Logit(y, X).fit()

params = model.params
conf = model.conf_int()
conf['Odds Ratio'] = params
conf.columns = ['5%', '95%', 'Odds Ratio']
print(np.exp(conf))
```
